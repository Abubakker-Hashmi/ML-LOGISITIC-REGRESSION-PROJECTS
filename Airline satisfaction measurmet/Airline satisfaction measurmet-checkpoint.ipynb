```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
Airline=pd.read_csv('Invistico_Airline.csv')
Airline.drop(['Customer Type','Type of Travel'],axis=1,inplace=True)
```

```python
sns.heatmap(Airline.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
mean_value = Airline['Arrival Delay in Minutes'].mean()
Airline['Arrival Delay in Minutes'].fillna(value=mean_value, inplace=True)
```

```python
class_mapping = {'Eco': 1, 'Eco Plus': 2, 'Business': 3}
Airline['Class'] = Airline['Class'].replace(class_mapping)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(Airline.drop('satisfaction',axis=1), 
                                                    Airline['satisfaction'], test_size=0.40, 
                                                    random_state=101) 
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

# Create a pipeline with scaling and logistic regression
ml = make_pipeline(StandardScaler(), LogisticRegression())
```

```python
ml.fit(X_train,y_train)
```

```python
y_pred=ml.predict(X_test)
len(y_pred)
```

```python
y_pred
```

```python
#Now checking accuracy of our model 
```

```python
plt.scatter(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_pred,y_test))
```

```python
cf=confusion_matrix(y_test,y_pred)
cf
```

```python
accuracy = ml.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```
